```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Some super basic stuff:**

```python
x = np.linspace(0,10,1000)
y = np.cos(x)

plt.plot(x,y)
plt.show()
```

**A little more jazz:**

```python
fig = plt.figure(1)
ax = fig.add_subplot(111)

ax.set_ylabel('cos(x)')
ax.set_xlabel('x')
ax.set_title('Super sweet plot')

plt.plot(x,y)
plt.show()
```

**Using grids:**

```python
fig = plt.figure(2)
ax = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax.plot(x,y)

y2 = np.sin(x)
ax2.plot(y2)

plt.show()
```

**A better back end in notebooks:**

```python
%matplotlib nbagg
```

```python
fig = plt.figure(3)
ax = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax.plot(x,y)

y2 = np.sin(x)
ax2.plot(y2)

plt.tight_layout(pad=1)

plt.show()
```

**Two y axes:**

```python
%matplotlib inline
```

```python
fig = plt.figure(4)
ax = fig.add_subplot(111)
ax2 = ax.twinx()

ax.plot(x,y)
ax2.plot(x,x, color = 'r')

plt.show()
```

**Pandas <--> Matplotlib**

```python
a = np.random.normal(0,1,100)
b = np.random.normal(0,1,100)
c = np.random.normal(0,1,100)
```

```python
df = pd.DataFrame({'a': a,
                   'b': b,
                   'c': c})
```

```python
df.head()
```

**Plot stuff directly:**

```python
plot = df.plot('a','b', kind = 'scatter')
```

**If you want to edit stuff, create figure/axes object first:**

```python
fig, ax = plt.subplots()

plot = df.plot('a','b', kind = 'scatter', ax = ax)

title = ax.set_title('Britannia Rulez')
ylabel = ax.set_ylabel('$10^{10^{10^{10^{10^{10^{10}}}}}}$')
xlabel = ax.set_xlabel('this bit normal    $abcd...xyz$')
```

**And a lil' sumpin' sumpin' for EDA:**
